```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from scipy.stats import ttest_ind

```

```python
df = pd.read_excel("C:/Users/Janhavi/Downloads/MASTERSHEET.xlsx")
```

```python
# Split the DataFrame into vegetarian and non-vegetarian groups
vegetarian_df = df[df['VEGETARIAN'] == 'Y']
non_vegetarian_df = df[df['VEGETARIAN'] == 'N']

# Conduct t-tests
ttest_results = {}

for column in ['BMI', 'WHR', 'SYSTOLIC BP', 'DIASTOLIC BP']:
    t_stat, p_value = ttest_ind(vegetarian_df[column], non_vegetarian_df[column])
    ttest_results[column] = {'t_statistic': t_stat, 'p_value': p_value}

# Visualize t-test results
for parameter, results in ttest_results.items():
    print(f"T-test results for {parameter}:")
    print(f"   - T-statistic: {results['t_statistic']}")
    print(f"   - P-value: {results['p_value']}")
    
# Calculate average systolic and diastolic BP for each group
average_bp = df.groupby('VEGETARIAN').agg({'SYSTOLIC BP': 'mean', 'DIASTOLIC BP': 'mean'}).reset_index()
average_bp = df.groupby('NON-VEGETARIAN').agg({'SYSTOLIC BP': 'mean', 'DIASTOLIC BP': 'mean'}).reset_index()

# Create separate bar charts for vegetarian and non-vegetarian groups
plt.figure(figsize=(12, 6))

# Bar chart for vegetarian group
plt.subplot(1, 2, 1)
sns.barplot(x='VEGETARIAN', y='SYSTOLIC BP', data=df, ci=None)
plt.title('Average Systolic BP for Vegetarian Group')

# Bar chart for non-vegetarian group
plt.subplot(1, 2, 2)
sns.barplot(x='NON-VEGETARIAN', y='DIASTOLIC BP', data=df, ci=None)
plt.title('Average Diastolic BP for Non-Vegetarian Group')

plt.tight_layout()
plt.show()    

```
